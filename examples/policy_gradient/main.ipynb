```python
import numpy as np
import matplotlib.pyplot as plt


def smooth_curve(y):
    N = len(y)
    x = np.arange(1, N+1)
    poly = np.polyfit(x, y, deg=10)
    return np.poly1d(poly)(x)

AC_returns = np.load('logs/returns_ActorCritic.npy').item()['returns']
REINFORCE_returns = np.load('logs/returns_REINFORCE.npy').item()['returns']

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(6, 2*4))

# Raw
axes[0].plot(x, REINFORCE_returns, label='REINFORCE')
axes[0].plot(x, AC_returns, label='Actor-critic')

# Smooth
axes[1].plot(x, smooth_curve(REINFORCE_returns), label='REINFORCE')
axes[1].plot(x, smooth_curve(AC_returns), label='Actor-critic')


for ax in axes:
    ax.set_title('Policy gradient on Acrobot-v1')
    ax.legend()

fig
```
